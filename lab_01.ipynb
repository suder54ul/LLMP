```python
import subprocess
import threading
import time

def ollama_serve():
    subprocess.Popen(["ollama", "serve"])

# Start the server in a separate thread
thread = threading.Thread(target=ollama_serve)
thread.start()
time.sleep(5) # Allow the server to initialize
```
